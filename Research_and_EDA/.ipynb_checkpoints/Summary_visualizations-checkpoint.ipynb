```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = '../Assets/merged.csv'
merged = pd.read_csv(path)
merged['Date'] = pd.to_datetime(merged.Date)
merged.set_index('Date', inplace=True)
```

```python
sns.set_context("notebook", font_scale=1.6)
sns.set_style("ticks")
sns.lmplot('Longitude', 'Latitude',
           data=merged, 
           fit_reg=False,
           hue="Station",
           scatter_kws={"marker": "D", 
                        "s": 100})
plt.title('Weather Station Clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
area = 900
colors = ['c','y']
plt.scatter([-87.933, -87.752], [41.995,41.786], 
            s=area, c = colors, marker='o', alpha=.6)
plt.figure(figsize=(12, 10))
```

```python
sns.set_context("notebook", font_scale=1.8)
sns.set_style("ticks")
sns.lmplot('Longitude', 'Latitude',
           data=merged, 
           fit_reg=False,
           hue="WnvPresent", col='Year', legend_out=True, scatter_kws={"marker": "D", "s": 100})
plt.title('WNV Geographic Dispersion')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.figure(figsize=(6, 6))
```
